```python
import pandas

order_reviews = pandas.read_csv("./olist_order_reviews_dataset.csv")
order_reviews.info()
order_reviews.head()
```

```python
review_text = order_reviews.dropna(subset=['review_comment_message']).copy()

print(f"Total de comentários reais para analisar: {len(review_text)}")
review_text['review_comment_message']
```

```python
rates = review_text['review_score'].value_counts().sort_index()
print("Distribuição das notas nos comentários:")
print(rates)
```

```python
orders = pandas.read_csv("./olist_orders_dataset.csv")
print(orders.info())
orders.head()
```

```python
order_consolided = pandas.merge(review_text,orders, on = 'order_id', how = 'left')

order_consolided['order_purchase_timestamp'] = pandas.to_datetime(order_consolided['order_purchase_timestamp'])
order_consolided['order_delivered_customer_date'] = pandas.to_datetime(order_consolided['order_delivered_customer_date'])
order_consolided['order_estimated_delivery_date'] = pandas.to_datetime(order_consolided['order_estimated_delivery_date'])

order_consolided['delivery_time'] = (order_consolided['order_delivered_customer_date'] - order_consolided['order_purchase_timestamp']).dt.days

print("Tabela unida com sucesso!")
order_consolided[['review_score', 'review_comment_message', 'delivery_time']].head()
```

```python
print("Média de dias para entrega por nota:")
print(order_consolided.groupby('review_score')['delivery_time'].mean())
```

```python
order_consolided['delivery_delay'] = (order_consolided['order_delivered_customer_date'] - order_consolided['order_estimated_delivery_date']).dt.days
print("Média de dias de ATRASO por nota (números positivos significam atraso real):")
print(order_consolided.groupby('review_score')['delivery_delay'].mean())
```

```python
review_1score = order_consolided[order_consolided['review_score'] == 1].copy()
review_1score_30c = review_1score[review_1score['review_comment_message'].str.len() > 30]
final_1score = review_1score_30c['review_comment_message'].dropna()

print(f"Temos {len(final_1score)} comentários detalhados de nota 1 com mais de 30 caracteres para a IA analisar.")

final_1score.to_excel('review_1score.xlsx') 
```

```python

import seaborn as sea
import matplotlib.pyplot as plt

df_analisado = pandas.read_csv('detractor_analysis_v2.csv')

# 2. Configurar o estilo visual
sea.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# 3. Criar o gráfico de contagem (Countplot)
# O 'order' garante que a maior barra fique no topo (mais organizado)
grafico = sea.countplot(
    data=df_analisado, 
    y='Category', 
    order=df_analisado['Category'].value_counts().index,
    palette='viridis',
    hue='Category',
    legend=False
)

# 4. Adicionar os números (rótulos) ao final de cada barra
for i in grafico.containers:
    grafico.bar_label(i, padding=3, fontsize=12, fontweight='bold')

# 5. Personalizar títulos e eixos
plt.title('Diagnóstico de Detratores: Principais Motivos de Nota 1', fontsize=16, pad=20)
plt.xlabel('Quantidade de Ocorrências', fontsize=12)
plt.ylabel('Categoria Semântica', fontsize=12)

# Ajustar o layout para não cortar os nomes
plt.tight_layout()

# Exibir
plt.show()
```
